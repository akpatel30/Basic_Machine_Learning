```python
import pandas as pd
```

```python
df=pd.read_csv('Melbourne_housing_FULL.txt')
df.head()
```

```python
df.nunique()
```

```python
df1=df.drop(['Address','Date','Postcode','YearBuilt','Lattitude','Longtitude'], axis='columns')
df1.head()
```

```python
cols_to_fill_zero=['Propertycount', 'Distance', 'Bedroom2', 'Bathroom', 'Car']
df1[cols_to_fill_zero]=df1[cols_to_fill_zero].fillna(0)
```

```python
df1.isnull().sum()
```

```python
import numpy as np
df1['Landsize']=df1.Landsize.fillna(df1.Landsize.mean())
df1.isnull().sum()
```

```python
df1['BuildingArea']=df1.BuildingArea.fillna(df1.BuildingArea.mean())
df1
```

```python
df1.shape
```

```python
df1.dropna(inplace=True)
df1
```

```python
df1.shape
```

```python
dataset=pd.get_dummies(df1, drop_first=True)
dataset
```

```python
Y=dataset.Price
X=dataset.drop(['Price'], axis='columns')
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test=train_test_split(X,Y, test_size=0.3, random_state=2)
```

```python
len(X_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
model.score(X_test,Y_test)
```

```python
model.score(X_train,Y_train)
```

## Lasso (L1 Regularized) Regression Model

```python
from sklearn.linear_model import Lasso
```

```python
L1_model=Lasso(alpha=50, max_iter=100, tol=0.1)
```

```python
L1_model.fit(X_train, Y_train)
```

```python
L1_model.score(X_test,Y_test)
```

```python
L1_model.score(X_train,Y_train)
```

## Ridge (L2 Regularized) Regression Model

```python
from sklearn.linear_model import Ridge
```

```python
L2_model=Ridge(alpha=50, max_iter=100, tol=0.1)
```

```python
L2_model.fit(X_train, Y_train)
```

```python
L2_model.score(X_test, Y_test)
```

```python
L2_model.score(X_train, Y_train)
```

### We see that Lasso and Ridge Regularizations prove to be beneficial when our Simple Linear Regression Model overfits. These results may not be that contrast but significant in most cases.Also that L1 & L2 Regularizations are used in Neural Networks too
